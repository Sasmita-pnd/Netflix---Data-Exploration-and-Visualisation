### Un-nesting the columns

```python
director = netflix_titles['director'].str.split(', ', expand=True).stack()
director
```

```python
director = director.reset_index(level=1, drop=True).to_frame('director')
director
```

```python
director['show_id'] = netflix_titles['show_id']
director
```

```python
cast = netflix_titles['cast'].str.split(', ', expand=True).stack()
cast =cast.reset_index(level=1, drop=True).to_frame('cast')
cast['show_id'] =netflix_titles['show_id']
cast
```

```python
country = netflix_titles['country'].str.split(', ', expand=True).stack()
country = country.reset_index(level=1, drop=True).to_frame('country')
country['show_id'] = netflix_titles['show_id']
country
```

```python
listed_in =netflix_titles['listed_in'].str.split(', ', expand=True).stack()
listed_in = listed_in.reset_index(level=1, drop=True).to_frame('listed_in')
listed_in['show_id'] = netflix_titles['show_id']
listed_in
```

```python
netflix_titles=netflix_titles.merge(director,on="show_id")
netflix_titles
```

```python
netflix_titles=netflix_titles.merge(cast,on="show_id")
netflix_titles
```

```python
netflix_titles=netflix_titles.merge(country,on="show_id")
netflix_titles
```

```python
netflix_titles=netflix_titles.merge(listed_in,on="show_id")
netflix_titles
```

```python
netflix_titles=netflix_titles.drop(["director_x","cast_x","country_x","listed_in_x"],axis=1)
```

```python
netflix_titles
```
